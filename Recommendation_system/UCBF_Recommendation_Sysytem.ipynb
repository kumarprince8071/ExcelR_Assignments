# Data Preprocesing

```python
book_data=book_data.rename(columns={'Unnamed: 0':'Unnamed','User.ID':'User_ID','Book.Title':'Book_title','Book.Rating':'Book_rating'})
```

```python
book_data.head()
```

```python
#number of unique users in the dataset
len(book_data.User_ID.unique())
```

```python
len(book_data.Book_title.unique())
```

```python
del book_data['Unnamed']
```

```python
book_data.head()
```

```python
user_book_df = pd.pivot_table(data = book_data,index='User_ID',
                                 columns='Book_title',
                                 values='Book_rating')#.reset_index(drop=True)
```

```python
user_book_df.head(20)
```

```python
user_book_df.index = book_data.User_ID.unique()
```

```python
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
