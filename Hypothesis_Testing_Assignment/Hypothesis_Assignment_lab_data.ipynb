### Separating Different features for statistical Test

```python
lab_1=lab_data['Laboratory 1']
lab_2=lab_data['Laboratory 2']
lab_3=lab_data['Laboratory 3']
lab_4=lab_data['Laboratory 4']
```

### Null HYpothesis- There is difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list
### Alternate Hypothesis - There is no difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list

# Performimg Ststistical Tests
### Here we have multiple samples of Quantitative type data so we are performing ANOVA Test

```python
from scipy.stats import f_oneway
```

```python
test_score,p_value=f_oneway(lab_1,lab_2,lab_3,lab_4)
```

```python
print('test_score {}\np_value {}.'.format(test_score,p_value))
```

```python
if p_value<0.05:
    print('There is no difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list')
else:
    print('There is difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list')
```
